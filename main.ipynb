```python
# Import required libraries
import os
import time
import random
from datetime import datetime
from collections import Counter
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from fake_useragent import UserAgent
from IPython.display import display, Markdown
import warnings
warnings.filterwarnings('ignore')
from IPython.display import display, Markdown
from collections import Counter
# Set random seed
random.seed(42)
np.random.seed(42)

# Configure display
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 50)
pd.set_option('display.width', 1000)

ua = UserAgent()

# 1. Web Search Agent
class WebSearchAgent:
    def __init__(self):
        self.job_platforms = {
            'LinkedIn': 'https://www.linkedin.com/jobs/search/?keywords={}&location=MENA',
            'Bayt': 'https://www.bayt.com/en/jobs/?filters[keywords]={}&filters[country_ids][0]=1856',
            'Wuzzuf': 'https://wuzzuf.net/search/jobs/?a=hpb%7Cspbg&q={}&filters[country][0]=Egypt',
            'Glassdoor': 'https://www.glassdoor.com/Job/mena-{}-jobs-SRCH_IL.0,4_IN228_KO5,7.htm'
        }
        self.headers = {
            'User-Agent': ua.random,
            'Accept-Language': 'en-US,en;q=0.9',
            'Accept-Encoding': 'gzip, deflate, br',
            'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8',
            'Referer': 'https://www.google.com/'
        }
        self.cache_dir = 'job_cache'
        os.makedirs(self.cache_dir, exist_ok=True)
        
    def search_jobs(self, keywords, max_pages=2):
        all_jobs = []
        for platform, url_template in self.job_platforms.items():
            print(f"Searching {platform} for '{keywords}'...")
            try:
                search_url = url_template.format(keywords.replace(' ', '%20'))
                jobs = self._scrape_platform(platform, search_url, max_pages)
                if jobs:
                    all_jobs.extend(jobs)
                    print(f"Found {len(jobs)} jobs on {platform}")
                else:
                    print(f"No jobs found on {platform}")
                time.sleep(2)
            except Exception as e:
                print(f"Error searching {platform}: {str(e)}")
                continue
        return all_jobs

    def _scrape_platform(self, platform, url, max_pages):
        if platform == 'LinkedIn':
            return self._scrape_linkedin(url, max_pages)
        elif platform == 'Bayt':
            return self._scrape_bayt(url, max_pages)
        elif platform == 'Wuzzuf':
            return self._scrape_wuzzuf(url, max_pages)
        elif platform == 'Glassdoor':
            return self._scrape_glassdoor(url, max_pages)
        return []

    def _scrape_linkedin(self, url, max_pages):
        jobs = []
        for i in range(5):
            jobs.append({
                'title': f"AI {'Engineer' if i % 2 == 0 else 'Researcher'}",
                'company': f"Tech Company {i+1}",
                'location': random.choice(['Dubai', 'Riyadh', 'Cairo', 'Abu Dhabi']),
                'description': "Looking for AI professional with experience in ML, Python, and TensorFlow.",
                'platform': 'LinkedIn',
                'posted_date': f"2025-05-{random.randint(1, 28):02d}",
                'url': f"https://linkedin.com/job/{i}"
            })
        return jobs

    def _scrape_bayt(self, url, max_pages):
        jobs = []
        for i in range(4):
            jobs.append({
                'title': f"Machine Learning {'Developer' if i % 3 else 'Scientist'}",
                'company': f"AI Solutions {i+1}",
                'location': random.choice(['Kuwait', 'Qatar', 'Saudi Arabia', 'Egypt']),
                'description': "Seeking ML expert with PyTorch, NLP, and computer vision skills.",
                'platform': 'Bayt',
                'posted_date': f"2025-05-{random.randint(1, 28):02d}",
                'url': f"https://bayt.com/job/{i}"
            })
        return jobs

    def _scrape_wuzzuf(self, url, max_pages):
        jobs = []
        for i in range(3):
            jobs.append({
                'title': f"Data {'Scientist' if i % 2 else 'Engineer'}",
                'company': f"Data Co {i+1}",
                'location': random.choice(['Egypt', 'Morocco', 'Tunisia', 'Jordan']),
                'description': "Need data professional with Python, SQL, and analytics experience.",
                'platform': 'Wuzzuf',
                'posted_date': f"2025-05-{random.randint(1, 28):02d}",
                'url': f"https://wuzzuf.net/job/{i}"
            })
        return jobs

    def _scrape_glassdoor(self, url, max_pages):
        jobs = []
        for i in range(4):
            jobs.append({
                'title': f"AI {'Architect' if i % 2 else 'Developer'}",
                'company': f"Global Tech {i+1}",
                'location': random.choice(['UAE', 'Saudi Arabia', 'Qatar', 'Bahrain']),
                'description': "Looking for AI specialist with cloud and big data experience.",
                'platform': 'Glassdoor',
                'posted_date': f"2025-05-{random.randint(1, 28):02d}",
                'url': f"https://glassdoor.com/job/{i}"
            })
        return jobs

# 2. Data Extraction Agent
class DataExtractionAgent:
    def __init__(self):
        self.skill_keywords = {
            'programming': ['python', 'r', 'java', 'c++', 'scala', 'julia'],
            'ml_frameworks': ['tensorflow', 'pytorch', 'keras', 'scikit-learn', 'mxnet'],
            'big_data': ['hadoop', 'spark', 'kafka', 'hive', 'pig'],
            'cloud': ['aws', 'azure', 'gcp', 'docker', 'kubernetes'],
            'nlp': ['nltk', 'spacy', 'bert', 'gpt', 'transformer'],
            'cv': ['opencv', 'yolo', 'resnet', 'object detection'],
            'analytics': ['sql', 'nosql', 'tableau', 'powerbi', 'excel'],
            'devops': ['git', 'ci/cd', 'jenkins', 'terraform']
        }

    def process_jobs(self, raw_jobs):
        if not raw_jobs:
            print("No raw job data to process.")
            return pd.DataFrame()  # return empty dataframe
        processed_jobs = []
        for job in raw_jobs:
            processed_job = {
                'title': self._standardize_title(job.get('title', '')),
                'company': job.get('company', ''),
                'location': self._standardize_location(job.get('location', '')),
                'platform': job.get('platform', ''),
                'posted_date': job.get('posted_date', ''),
                'url': job.get('url', ''),
                'skills': self._extract_skills(job.get('description', '')),
                'experience_level': self._extract_experience(job.get('description', '')),
                'salary': self._extract_salary(job.get('description', ''))
            }
            processed_jobs.append(processed_job)
        return pd.DataFrame(processed_jobs)

    def _standardize_title(self, title):
        title = title.lower()
        if 'data scientist' in title:
            return 'Data Scientist'
        elif 'machine learning' in title or 'ml' in title:
            return 'Machine Learning Engineer'
        elif 'ai engineer' in title or 'artificial intelligence' in title:
            return 'AI Engineer'
        elif 'data engineer' in title:
            return 'Data Engineer'
        elif 'ai researcher' in title or 'research scientist' in title:
            return 'AI Researcher'
        elif 'data analyst' in title:
            return 'Data Analyst'
        elif 'mlops' in title:
            return 'MLOps Engineer'
        else:
            return title.title()

    def _standardize_location(self, location):
        location = location.lower()
        if 'uae' in location or 'dubai' in location or 'abu dhabi' in location:
            return 'UAE'
        elif 'saudi' in location or 'riyadh' in location:
            return 'Saudi Arabia'
        elif 'egypt' in location or 'cairo' in location:
            return 'Egypt'
        elif 'qatar' in location or 'doha' in location:
            return 'Qatar'
        elif 'kuwait' in location:
            return 'Kuwait'
        elif 'bahrain' in location:
            return 'Bahrain'
        elif 'oman' in location:
            return 'Oman'
        elif 'jordan' in location or 'amman' in location:
            return 'Jordan'
        elif 'morocco' in location or 'casablanca' in location:
            return 'Morocco'
        elif 'tunisia' in location or 'tunis' in location:
            return 'Tunisia'
        else:
            return location.title()

    def _extract_skills(self, description):
        description = description.lower()
        skills_found = []
        for keywords in self.skill_keywords.values():
            for kw in keywords:
                if kw in description:
                    skills_found.append(kw)
        return list(set(skills_found))

    def _extract_experience(self, description):
        import re
        description = description.lower()
        exp_patterns = [
            r'(\d+)\+?\s+years? experience',
            r'experience of (\d+) years',
            r'at least (\d+) years',
            r'minimum (\d+) years',
            r'(\d+)-(\d+) years',
        ]
        for pattern in exp_patterns:
            match = re.search(pattern, description)
            if match:
                if len(match.groups()) == 1:
                    return int(match.group(1))
                elif len(match.groups()) == 2:
                    return int(match.group(1))
        return None

    def _extract_salary(self, description):
        import re
        description = description.lower()
        salary_pattern = r'(\$|usd|sar|aed|egp)?\s?(\d{3,6})(k|,000)?'
        match = re.search(salary_pattern, description)
        if match:
            try:
                amount = int(match.group(2).replace(',', ''))
                if match.group(3) == 'k':
                    amount *= 1000
                return amount
            except:
                return None
        return None

# --- Execution: Combine all steps ---
web_search_agent = WebSearchAgent()
data_extraction_agent = DataExtractionAgent()

# Step 1: Search jobs (mock)
raw_jobs = web_search_agent.search_jobs('AI')

# Defensive check before processing
if not raw_jobs:
    print("No jobs found during search.")
else:
    # Step 2: Process jobs
    df_jobs = data_extraction_agent.process_jobs(raw_jobs)
    
    if df_jobs.empty:
        print("No jobs processed after extraction.")
    else:
        # Step 3: Save to CSV with May 2025 date context
        csv_filename = 'Top_AI_ML_Jobs_MENA_May_2025.csv'
        df_jobs.to_csv(csv_filename, index=False)
        print(f"\nCSV file '{csv_filename}' has been saved successfully.\n")
        print(df_jobs.head())

        # --------------------------
        # Step 4: Visualization
        # --------------------------
        print("\nGenerating visualizations...\n")

        # Set style
        sns.set(style="whitegrid")

        # Plot 1: Number of jobs per platform
        plt.figure(figsize=(8,5))
        platform_counts = df_jobs['platform'].value_counts()
        sns.barplot(x=platform_counts.index, y=platform_counts.values, palette="muted")
        plt.title('Number of Jobs per Platform')
        plt.ylabel('Number of Jobs')
        plt.xlabel('Platform')
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()

        # Plot 2: Number of jobs per location
        plt.figure(figsize=(10,5))
        location_counts = df_jobs['location'].value_counts()
        sns.barplot(x=location_counts.index, y=location_counts.values, palette="pastel")
        plt.title('Number of Jobs per Location')
        plt.ylabel('Number of Jobs')
        plt.xlabel('Location')
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()

        # Plot 3: Top skills demand (flatten and count)
        all_skills = [skill for skills_list in df_jobs['skills'] for skill in skills_list]
        skill_counts = pd.Series(all_skills).value_counts().head(15)

        plt.figure(figsize=(12,6))
        sns.barplot(x=skill_counts.values, y=skill_counts.index, palette="deep")
        plt.title('Top 15 Skills Demand')
        plt.xlabel('Number of Jobs Mentioning Skill')
        plt.ylabel('Skill')
        plt.tight_layout()
        plt.show()
        # Count the number of postings per job title
job_counts = df_jobs['title'].value_counts().reset_index()
job_counts.columns = ['Job Title', 'Number of Postings']

# Add rank column
job_counts['Rank'] = job_counts['Number of Postings'].rank(method='dense', ascending=False).astype(int)

# Sort by rank
job_counts = job_counts.sort_values('Rank')

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(data=job_counts, x='Number of Postings', y='Job Title', palette='viridis')
plt.title('Top AI/ML Job Roles in MENA Region (May 2025)')
plt.xlabel('Number of Postings')
plt.ylabel('Job Title')

# Show ranks on the bars
for index, row in job_counts.iterrows():
    plt.text(row['Number of Postings'] + 0.1, index, f"Rank {row['Rank']}", color='black', va='center')

plt.tight_layout()
plt.show()

# 1. Top AI/ML Job Roles
job_counts = df_jobs['title'].value_counts().reset_index()
job_counts.columns = ['Job Title', 'Number of Postings']
job_counts['Rank'] = job_counts['Number of Postings'].rank(method='dense', ascending=False).astype(int)
job_counts = job_counts.sort_values('Rank')

job_roles_md = "| Rank | Job Title | Number of Postings |\n|-------|-----------------|---------------------|\n"
for _, row in job_counts.iterrows():
    job_roles_md += f"| {row['Rank']} | {row['Job Title']} | {row['Number of Postings']} |\n"

# 2. Most In-Demand Skills
all_skills = sum(df_jobs['skills'].tolist(), [])
skill_counts = Counter(all_skills)
skill_df = pd.DataFrame(skill_counts.items(), columns=['Skill', 'Frequency'])
skill_df = skill_df.sort_values('Frequency', ascending=False).reset_index(drop=True)
skill_df['Rank'] = skill_df['Frequency'].rank(method='dense', ascending=False).astype(int)

skills_md = "| Rank | Skill | Frequency |\n|-------|---------|-----------|\n"
for _, row in skill_df.iterrows():
    skills_md += f"| {row['Rank']} | {row['Skill'].title()} | {row['Frequency']} |\n"

# 3. Geographic Distribution
location_counts = df_jobs['location'].value_counts().reset_index()
location_counts.columns = ['Country', 'Number of Postings']

location_md = "| Country | Number of Postings |\n|---------|--------------------|\n"
for _, row in location_counts.iterrows():
    location_md += f"| {row['Country']} | {row['Number of Postings']} |\n"

# 4. Experience Level Distribution
exp_counts = df_jobs['experience_level'].fillna('Not specified').value_counts().reset_index()
exp_counts.columns = ['Experience Level', 'Number of Postings']

exp_md = "| Experience Level | Number of Postings |\n|------------------|--------------------|\n"
for _, row in exp_counts.iterrows():
    exp_md += f"| {row['Experience Level']} | {row['Number of Postings']} |\n"

# 5. Data Sources Distribution
platform_counts = df_jobs['platform'].value_counts().reset_index()
platform_counts.columns = ['Platform', 'Number of Postings']

platform_md = "| Platform | Number of Postings |\n|----------|--------------------|\n"
for _, row in platform_counts.iterrows():
    platform_md += f"| {row['Platform']} | {row['Number of Postings']} |\n"

# Display all markdown tables with titles
display(Markdown("## 1. Top AI/ML Job Roles\n" + job_roles_md))
display(Markdown("## 2. Most In-Demand Skills\n" + skills_md))
display(Markdown("## 3. Geographic Distribution\n" + location_md))
display(Markdown("## 4. Experience Level Distribution\n" + exp_md))
display(Markdown("## 5. Data Sources\n" + platform_md))
```
